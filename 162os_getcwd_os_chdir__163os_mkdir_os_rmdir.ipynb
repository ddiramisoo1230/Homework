```python
#162
from os import getcwd
print(f'현재 작업 디렉토리:{getcwd()}')
```

```python
from os import chdir
print(f'현재 작업 디렉토리:{getcwd()}')
chdir('..')
print(f'현재 작업 디렉토리: {getcwd()}')
chdir('c:/mywork/jupyter')
print(f'현재 작업 디렉토리: {getcwd()}')
```

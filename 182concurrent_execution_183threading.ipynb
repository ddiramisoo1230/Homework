```python
#183
from time import time

def checkPrimeNumber(x):
  ret=False

  if x==2:
    return True
  if x<=1 or x%2==0:
    return False

  print(f'{x:,}가 소수인지 아닌지 판별 중...')
  stime =time()
  for i in range(3, int(x**(1/2))+1,2):
    if x%i == 0:
      ret = False
  else:
    ret=True
  etime = time()

  if ret:
    print(f'{x:,}는 소수입니다. 소요 시간: {etime-stime:.2f}초')
  else:
    print(f'{x:,}는 소수가 아닙니다.')

  return ret
```

```python
def cumulativeSum(n):
  ret=0
  print(f'{n:,}까지 누적합 계산중...')
  stime = time()
  for i in range(1,n+1):
    ret+=i
  etime = time()
  print(f'{n:,}까지 누적합 계산 시간 {etime-stime:2f}초')
  print(f'누적합 결과: {ret:,}')
  return ret
```

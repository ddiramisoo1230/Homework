```python
#172
from time import localtime, strftime
weekdays=['월요일','화요일','수요일','목요일','금요일','토요일','일요일']
t=localtime()
today=strftime('%Y-%m-%d',t)
print(f'{today}은 {weekdays[t.tm_wday]} 입니다.')
```

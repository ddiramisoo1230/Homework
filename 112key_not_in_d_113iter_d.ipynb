```python
#112
frequency={}
while True:
  c=input('알파벳 1자를 입력하세요[끝내려면 엔터를 치세요]:')
  if c=='':
    break
  if c not in frequency:
    frequency[c]=1
  else:
    frequency[c]+=1
  print(frequency)
```

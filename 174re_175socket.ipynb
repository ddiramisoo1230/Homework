```python
#174
import re
p=re.compile(r'href=".*"',re.I)
```

```python
contents="""<a href="help.html">도움말을 보려면 여기를 클릭하세요</a><fontsize="15">"""
ret=p.match(contents)
print(ret)
```

```python
contents2="""href="help.html">도움말을 보려면 여기를 클릭하세요</a><fontsize="15">"""
ret=p.match(contents2)
print(ret.group())
```

```python
ret=p.search(contents)
print(f'Match.group()={ret.group()}')
print(f'Match.start()={ret.start()}')
print(f'Match.end()={ret.end()}')
print(f'Match.span()={ret.span()}')
```

```python
ret=p.findall(contents)
print(ret)
```

```python
#175
import socket
HOST=''
PORT=7777
def echoServer():
  with socket.socket(socket.AF_INEF,socket.SOCK_STREAM) as sock:
    sock.bind((HOST,PORT))
    sock.listen()
    print('클라이언트 연결을 기다리는 중...')
    conn, addr= sock.accept()
    with conn:
      print(f'{addr[0]}와 연결됨')
      while True:
        data=conn.recv(1024)
        if not data:
          break
          print(f'수신 메세지: {data.decode()}')
          conn.sendall(data)
echoServer()
```
